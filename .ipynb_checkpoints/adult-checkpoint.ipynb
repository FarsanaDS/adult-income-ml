## Data cleaning and Preprocessing

```python
# Check for Duplicates
data.duplicated().sum()
```

```python
# Remove duplicate rows
data = data.drop_duplicates()
```

```python
# Verify if duplicates are removed
data.duplicated().sum()
```

```python
# Fill missing values
data['workclass'] = data['workclass'].fillna(data['workclass'].mode()[0])
data['occupation'] = data['occupation'].fillna(data['occupation'].mode()[0])
data['native.country'] = data['native.country'].fillna(data['native.country'].mode()[0])
```

```python
# Cnfirm no missing values remain
data.isnull().sum()
```

```python
# Remove leading/trailing spaces from categorical values
for col in data.select_dtypes(include='object').columns:
    data[col] = data[col].str.strip()
```

```python
# Drop redundant column: 'education' (duplicate of 'education-num')
data.drop('education', axis=1, inplace=True)
```

```python
# Convert income to binary ( 0 for <=50K, 1 for >50K)
data['income'] = data['income'].map({'<=50K': 0, '>50K': 1})
```

```python
# Calculate interquartile range (IQR) for age
Q1 = data['age'].quantile(0.25)
Q3 = data['age'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR
```

```python
# Clip age column to remove outliers
data['age'] = data['age'].clip(lower=lower_bound, upper=upper_bound)
```

```python
# Boxplot to visualize outliers
sns.boxplot(data=data, y='age')
plt.show()
```

## Feature Engineering

#### Feature Expansion

```python
# Define Age groups
def ageGrouping(column):
    if 17 <= column <= 30:
        return 'Youth'
    elif 30 < column <= 50:
        return 'Middle-aged'
    else:
        return 'Senior'
data['ageGroup'] = data['age'].apply(ageGrouping)
```

```python
# Calculate net capital (gain - loss)
data['net'] = data['capital.gain'] - data['capital.loss']
```

```python
# Workload relative to age
data['work_hours_per_age'] = data['hours.per.week']/data['age']
```

```python
# Categorize hours worked per week
data['hours_category'] = pd.cut(data['hours.per.week'], bins=[0,30,40,60,100],
                               labels=['Part-Time','Full-Time','Over-Time','Extreme'])
```

```python
# Binary flag for capital activity
data['capital_activity'] = np.where((data['capital.gain'] > 0) | (data['capital.loss'] > 0),1,0)
```

```python
# Display the count of unique values in the 'marital.status' column
data['marital.status'].value_counts()
```

```python
# Define consolidated categories for marital.status
data['marital.status'] = data['marital.status'].replace({
    'Married-civ-spouse':'Married-spouse-present',
    'Married-AF-spouse':'Married-spouse-present',
    'Married-spouse-absent':'Absent',
    'Widowed':'Absent',
    'Divorced':'Divorced',
    'Separated':'No-partner',
    'Never-married':'No-partner'
})
```

```python
data['native.country'].value_counts()
```

```python
# Group 'native.country' into 'United-states' and 'Other'
data['native.country'] = data['native.country'].apply(lambda x: 'Other' if x != 'United-states' else x)
```

```python
data['occupation'].value_counts()
```

```python
# Group rare occupations as 'Other'
data['occupation'] = data['occupation'].apply(
    lambda x: x if data['occupation'].value_counts()[x] > 1000 else 'Other')
```

```python
# Ordinal encoding for ageGroup and hours_category
ageGroupOrder = ['Youth', 'Middle-aged', 'Senior']
hours_category_order = ['Part-Time', 'Full-Time', 'Over-Time', 'Extreme']
```

```python
ordinal_encoder = OrdinalEncoder(categories=[ageGroupOrder, hours_category_order])
data[['ageGroup', 'hours_category']] = ordinal_encoder.fit_transform(data[['ageGroup', 'hours_category']]).astype(int)
```

```python
# Onehot Encoding for categorical features
OneHotColumns = ['workclass', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']
data_encoded = pd.get_dummies(data[OneHotColumns], drop_first=True).astype(int)
```

```python
# Concatenate encoded columns back to the data frame
data = pd.concat([data.drop(columns=OneHotColumns), data_encoded], axis=1)
```

```python
# Display the dataset
data.head()
```

## Split The Dataset

```python
# Define features (x) and target variable (y)
x = data.drop('income', axis=1)
y = data['income']
```

```python
# Assuming x is originally a dataframe
x_columns = x.columns
# save original column names
```

## Feature Selection

```python
selector = SelectKBest(score_func=f_classif, k=13) # Use ANOVA F-test
x = selector.fit_transform(x,y)
selected_features = x_columns[selector.get_support()]
print('Top features (ANOVA):', selected_features.tolist())
```

```python
# List of selected features
selected_features_list = ['age', 'education.num', 'capital.gain', 'hours.per.week', 'ageGroup', 'net', 'hours_category', 'capital_activity',
                          'marital.status_Married-spouse-present',
                          'marital.status_No-partner', 'occupation_Exec-managerial', 'relationship_Own-child', 'sex_Male']
# Add the target column to the selected features
selected_features_list.append('income')
```

```python
# Filter the dataset to keep only the selected features and target
filtered_dataset = data[selected_features_list]

# Display the filtered dataset
filtered_dataset.head()
```

```python
# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Outlier Treatment

```python
# Cnvert the feature from the training set into a numpy array
a = np.array(x_train[0])
# Calculate the 25th percentile(Q1) and 75th percentile(Q3) of the feature 'a'
Q1 = np.percentile(a, 25)
Q3 = np.percentile(a, 75)

# Compute the interquartile range
IQR = Q3 - Q1
# Determine the lower and upper bounds for detecting outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print( f'Lower Bound: {lower_bound}, Upper Bound: {upper_bound}')

# Identify the outliers in the data
outliers = a[(a < lower_bound) | (a > upper_bound)]
print(f'Outliers: {outliers}')
print('Number of Outliers:', len(outliers))
print('Percentage of Outliers:',len(outliers)/a.shape[0]*100,'%')

# Remove outliers: Keep only values within the bounds
a_without_outliers = a[(a >= lower_bound) & (a <= upper_bound)]

# Print the cleaned data
print('Data without Outliers:', a_without_outliers)
print('Number of Data Points After Removing Outliers:', len(a_without_outliers))

# Visualize the cleaned feature using a boxplot
sns.boxplot(a_without_outliers)
plt.show()
```

## Handle Class Imbalance

```python
# Apply SMOTE to balace the dataset
smote = SMOTE(random_state=42, k_neighbors= 3, sampling_strategy=0.5)
x_train, y_train = smote.fit_resample(x_train, y_train)
```

```python
# Check the new class distribution
print('Class distribution after SMOTE:', np.bincount(y_train))
```

```python
len(x_train)
```

```python
y_train.value_counts()
```

```python
# Standardize the numerical features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

## Model building(GridSearchCV) and Evaluation

#### 1. Logistic Regression

```python
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga']
}

grid_search_lr = GridSearchCV(
    estimator=LogisticRegression(max_iter=1000, random_state=42),
    param_grid=param_grid_lr, scoring='accuracy', cv=5, verbose=1, n_jobs=-1
)

# Fit and evaluate
grid_search_lr.fit(x_train_scaled, y_train)
best_lr = grid_search_lr.best_estimator_

# Predictions
y_train_pred_lr = best_lr.predict(x_train_scaled)
y_test_pred_lr = best_lr.predict(x_test_scaled)

# Metrics
print("Logistic Regression Results:")
print("Best Parameters:", grid_search_lr.best_params_)
print("Train Accuracy:", accuracy_score(y_train, y_train_pred_lr))
print("Test Accuracy:", accuracy_score(y_test, y_test_pred_lr))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred_lr))
print("Classification Report:\n", classification_report(y_test, y_test_pred_lr))
```

#### 2. Decision Tree

```python
param_grid_dt = {
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search_dt = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=42),
    param_grid=param_grid_dt, scoring='accuracy', cv=5, verbose=1, n_jobs=-1
)

# Fit and evaluate
grid_search_dt.fit(x_train, y_train)
best_dt = grid_search_dt.best_estimator_

# Predictions
y_train_pred_dt = best_dt.predict(x_train)
y_test_pred_dt = best_dt.predict(x_test)

# Metrics
print("Decision Tree Results:")
print("Best Parameters:", grid_search_dt.best_params_)
print("Train Accuracy:", accuracy_score(y_train, y_train_pred_dt))
print("Test Accuracy:", accuracy_score(y_test, y_test_pred_dt))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred_dt))
print("Classification Report:\n", classification_report(y_test, y_test_pred_dt))
```

#### 3. Random Forest

```python
param_grid_rf = {
    'n_estimators': [50, 100, 200, 300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'bootstrap': [True, False]
}

grid_search_rf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid_rf, scoring='accuracy', cv=5, verbose=1, n_jobs=-1
)

# Fit and evaluate
grid_search_rf.fit(x_train, y_train)
best_rf = grid_search_rf.best_estimator_

# Predictions
y_train_pred_rf = best_rf.predict(x_train)
y_test_pred_rf = best_rf.predict(x_test)

# Metrics
print("Random Forest Results:")
print("Best Parameters:", grid_search_rf.best_params_)
print("Train Accuracy:", accuracy_score(y_train, y_train_pred_rf))
print("Test Accuracy:", accuracy_score(y_test, y_test_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_test_pred_rf))
```

#### 4. SVC

```python
param_grid_svc = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}

grid_search_svc = GridSearchCV(
    estimator=SVC(random_state=42),
    param_grid=param_grid_svc, scoring='accuracy', cv=5, verbose=1, n_jobs=-1
)

# Fit and evaluate
grid_search_svc.fit(x_train_scaled, y_train)
best_svc = grid_search_svc.best_estimator_

# Predictions
y_train_pred_svc = best_svc.predict(x_train_scaled)
y_test_pred_svc = best_svc.predict(x_test_scaled)

# Metrics
print("SVC Results:")
print("Best Parameters:", grid_search_svc.best_params_)
print("Train Accuracy:", accuracy_score(y_train, y_train_pred_svc))
print("Test Accuracy:", accuracy_score(y_test, y_test_pred_svc))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred_svc))
print("Classification Report:\n", classification_report(y_test, y_test_pred_svc))
```

#### 5. K-Nearest Neighbors (KNN)

```python
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance']
}

grid_search_knn = GridSearchCV(
    estimator=KNeighborsClassifier(),
    param_grid=param_grid_knn, scoring='accuracy', cv=5, verbose=1, n_jobs=-1
)

# Fit and evaluate
grid_search_knn.fit(x_train_scaled, y_train)
best_knn = grid_search_knn.best_estimator_

# Predictions
y_train_pred_knn = best_knn.predict(x_train_scaled)
y_test_pred_knn = best_knn.predict(x_test_scaled)

# Metrics
print("KNN Results:")
print("Best Parameters:", grid_search_knn.best_params_)
print("Train Accuracy:", accuracy_score(y_train, y_train_pred_knn))
print("Test Accuracy:", accuracy_score(y_test, y_test_pred_knn))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred_knn))
print("Classification Report:\n", classification_report(y_test, y_test_pred_knn))
```

#### 6. XGBoost

```python
param_grid_xgb = {
    'n_estimators': [50, 100, 150],
    'max_depth': [3, 6, 10],
    'learning_rate': [0.01, 0.1, 0.2]
}

grid_search_xgb = GridSearchCV(
    estimator=XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    param_grid=param_grid_xgb, scoring='accuracy', cv=5, verbose=1, n_jobs=-1
)

# Fit and evaluate
grid_search_xgb.fit(x_train, y_train)
best_xgb = grid_search_xgb.best_estimator_

# Predictions
y_train_pred_xgb = best_xgb.predict(x_train)
y_test_pred_xgb = best_xgb.predict(x_test)

# Metrics
print("XGBoost Results:")
print("Best Parameters:", grid_search_xgb.best_params_)
print("Train Accuracy:", accuracy_score(y_train, y_train_pred_xgb))
print("Test Accuracy:", accuracy_score(y_test, y_test_pred_xgb))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred_xgb))
print("Classification Report:\n", classification_report(y_test, y_test_pred_xgb))
```

#### 7. Gaussian Naive Bayes

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)

# Predictions
y_train_pred_gnb = gnb.predict(x_train)
y_test_pred_gnb = gnb.predict(x_test)

# Metrics
print("Gaussian Naive Bayes Results:")
print("Train Accuracy:", accuracy_score(y_train, y_train_pred_gnb))
print("Test Accuracy:", accuracy_score(y_test, y_test_pred_gnb))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred_gnb))
print("Classification Report:\n", classification_report(y_test, y_test_pred_gnb))
```
